```python
import pandas as pd
```

```python
df = pd.read_csv("Taxi_Trips_-_2019.tsv", header=0, delimiter="\t")
```

```python
df.columns.values.tolist()
```

```python
# 3.Trip Start Timestamp (string -> date and time)
# 5. Trip Seconds (int)
# 6. Trip Miles (float)
# 9. Pickup Community Area (int)
# 10. Drop-off community Area (int)
# 17. Company (string)
df = df[['Trip Start Timestamp', 'Trip Seconds', 'Trip Miles', 'Pickup Community Area', 'Dropoff Community Area', 'Company']]
df.head(5)
# keep only necessary columns
```

```python
df = df[df['Pickup Community Area'].notnull() & df['Dropoff Community Area'].notnull()]
df
#delete community areas NaN since those are outside of chicago.
#14629392 rows × 6 columns
```

```python
# remove all trips less than 0.5 miles, and more than 100 miles
# df[df['Trip Miles'] < 0.5]
df = df[(df['Trip Miles'] >= 0.5) & (df['Trip Miles'] <= 100)]
df
# 11410676 rows × 6 columns
```

```python
# remove less than 60 seconds, and greater than 5 hours (18,000 secs)
# df[df['Trip Seconds'] > 18000]
df = df[(df['Trip Seconds'] >= 60) & (df['Trip Miles'] <= 18000)]
df
# 11403451 rows × 6 columns
```

```python
df.to_csv("taxi_data_all.csv", index=False, header=True, mode='a')
```
